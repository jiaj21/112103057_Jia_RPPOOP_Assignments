```python
# TYPES OF INHERITANCE

import math

class polygon:
    def __init__(self,num_sides):
        self.num_sides=num_sides

    def shape(self):
        if self.num_sides==3:
            return "Triangle"
        elif self.num_sides==4:
            return "Square or Rectangle"
        else:
            return "polygon with sides more than 4"
   
    def area(self):
        pass
    def perimeter(self):
        pass
    def get_num_sides(self):
        return self.num_sides

# hierarchical inheritance
class rectangle(polygon):
    def __init__(self,l,b):
        polygon.__init__(self,4)
        self.length=l
        self.breadth=b
   
    def area(self):
        return self.length*self.breadth
   
    def perimeter(self):
        return 2*(self.length+self.breadth)
   
    def get_dimension(self):
        return self.length,self.breadth
   
    def set_dimension(self,l,b):
        self.length=l
        self.breadth=b
   
    def shape(self):
        return "Rectangle"

# hierarchical inheritance
class triangle(polygon):
    def __init__(self,a,b,c):
        polygon.__init__(self,3)
        self.a=a
        self.b=b
        self.c=c
   
    def area(self):
        s=(self.a+self.b+self.c)/2
        return (s*(s-self.a)*(s-self.b)*(s-self.c))**0.5
   
    def perimeter(self):
        return self.a+self.b+self.c
   
    def get_dimension(self):
        return self.a,self.b,self.c
   
    def set_dimension(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c


# multilevel inheritance
class square(rectangle):
    def __init__(self,a):
        rectangle.__init__(self,a,a)
        self.side=a
   
    def get_dimension(self):
        return self.side
   
    def set_dimension(self,a):
        self.side=a
       
    def shape(self):
        return "Square"

# single level inheritance 

class Ellipse:
    def __init__(self,a,b):
        self.a=a
        self.b=b

    def area(self):
        return math.pi*self.a*self.b
    def shape(self):
        return "Ellipse"
   
   
class Circle(Ellipse):
    def __init__(self,r):
        self.r=r
        Ellipse.__init__(self,r,r)
    def shape(self):
        return "Cirlce"
    
#multiple inheritance
class Father:
    fname="JOHNSON"
    
class Mother:
    mname="ROSE"

class Child(Father,Mother):
    def __init__(self,name):
        self.name=name
    def Chname(self):
        print(self.name)
    def Parents(self):
        print(self.fname,self.mname)
        
   

# object for square
sq=square(5)
print(sq.shape())
print(sq.get_dimension())
print(sq.area())
print(sq.perimeter())
sq.set_dimension(10)
print(sq.get_dimension())
print(sq.get_num_sides())
print()

#object for rectangle
rect=rectangle(3,4)
print(rect.shape())
print(rect.get_dimension())
print(rect.area())
print(rect.perimeter())
rect.set_dimension(10,12)
print(rect.get_dimension())
print(rect.get_num_sides())
print()

#object for triangle
tri=triangle(3,4,5)
print(tri.shape())
print(tri.get_dimension())
print(tri.area())
print(tri.perimeter())
tri.set_dimension(8,6,10)
print(tri.get_dimension())
print(tri.get_num_sides())
print()

#object for ellipse
ell=Ellipse(4,5)
print(ell.shape())
print(ell.area())
print()

#object for circle
cir=Circle(10)
print(cir.shape())
print(cir.area())
print()

#multiple inheritance
#object for father
ch=Child("Jia")
ch.Chname()
ch.Parents()
```
