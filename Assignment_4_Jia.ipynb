```python
from random import randint
    
class Customer():
    name=""
    balance_amount=0
    __password=0
    __account_number=0
    
    def __init__(self,name="",deposit=0):
        self.name=name
        self.balance_amount=deposit
        self.__account_number=randint(10000, 99999)
        
    def displayAccountNumber(self):
        return self.__account_number
        
    def setPassword(self,password):
        self.__password=password
        
    def withdrawAmount(self, amount):
        if self.balance_amount >= amount:
            self.balance_amount -= amount
            print()
            print("The sum of " ,amount, " has been withdrawn from your account balance.")
            self.balanceAmount()
        else:
            print()
            print("Not sufficient Balance!")
        
    
    def depositAmount(self, amount):
        self.balance_amount += amount
        print()
        print("The sum of ",amount," has been added to your account balance.")
        self.balanceAmount()
        
    def balanceAmount(self):
        print("Your current balance amount is", self.balance_amount)
        
    def authenticate(self):
            return self.__account_number,self.__password
        
class Bank(Customer):
    def __init__(self):
        Customer.__init__(self)
        
    bank_name="State Bank Of India"
    customer = []

    def update(self, customer):
        self.customer.append(customer)
    
    def authentication(self,account_num,passd):
        for i in self.customer:
            a,b=i.authenticate()
            if(account_num==a)&(passd==b):
                return i


#main program
bank=Bank()
print("Welcome to ",bank.bank_name," ! \n")
a=True
while a:
    print()
    choice = int(input("Choose an option:\n1. Open new bank account\n2. Open existing bank account\n3. Exit"))
    
    if choice == 1:
        print()
        print("Please fill in the information below.")
        print()
        cust = Customer(input("Name: "), int(input("Deposit amount: ")))
        bank.update(cust)
        print()
        print("Account created successfully! Your account number is: ", cust.displayAccountNumber())
        password=int(input("Enter Password : "))
        cust.setPassword(password)
    elif choice == 2:
        print()
        print("Please enter your credentials below.")
        print()
        acc_num = int(input("Account number : "))
        acc_password = int(input("Password : "))
        current_client = bank.authentication(acc_num,acc_password)
        if current_client:
            print()
            print("Welcome {}!".format(current_client.name))
            acc_open = True
            while acc_open:
                print()
                acc_choice = int(input("Choose an option: 1.Withdraw\n 2. Deposit\n 3. Balance\n 4. Exit"))
                if acc_choice == 1:
                    print()
                    current_client.withdrawAmount(int(input("Withdraw amount: ")))
                elif acc_choice == 2:
                    print()
                    current_client.depositAmount(int(input("Deposit amount: ")))
                elif acc_choice == 3:
                    print()
                    current_client.balanceAmount()
                elif acc_choice == 4:
                    print()
                    print("Thank you for visiting!")
                    current_client = ''
                    acc_open = False
        else:
            print()
            print("Authentication failed!")
            print("Reason: account not found.")
            continue
    elif choice == 3:
        print()
        print("Thank You!")
        a=False
    
            
        

    
    
        
    
```
